# Funcion de coste

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm # Para manejar colores
lm = LinearRegression()
```

```python
x = np.array([0,20,30,40,50,60])
y= np.array([100,150,200,180,250,230])

lm.fit(x.reshape(-1,1),y.reshape(-1,1))
print("Mi b_0 es ", lm.intercept_,' Y mi b_1= ',lm.coef_)
plt.scatter(x,y, color = 'b', marker = 'o', s = 30)
plt.plot(x,y, color= 'r')
plt.show()
sns.regplot(x.reshape(-1,1),y.reshape(-1,1))
```

```python
from sklearn.metrics import mean_squared_error

Yhat=lm.predict(x.reshape(-1,1))
print('The output of the first four predicted value is: ', Yhat)

mse = mean_squared_error(y.reshape(-1,1), Yhat)


print('The mean square error of price and predicted value is: ', mse)
```

# Graficando en 3D la funcion de coste

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

'''Definimos la funcion con la que vamos a trabajar'''
def f(x,y):
    return x**2 + y**2

'''Definimos los vectores'''
#x = np.linspace(-4,4,num=100)
#y = np.linspace(-4,4,num=100)

'''Generar nuestro meshgrid'''
X,Y = np.meshgrid(x,y)

Z =f(X,Y)


'''Crearemos nuestra superficie'''
surf = ax.plot_surface(X,Y,Z, cmap=cm.cool)

'''Creamos la linea de colores'''
fig.colorbar(surf)

```

# Curvas de nivel

```python
'''Graficar las curvas de nivel'''
fig2,ax2 = plt.subplots()

'''Graficar los niveles'''
level_map = np.linspace(np.min(Z),np.max(Z), num= 100)

'''Hacemos un countour plot'''
cp = ax2.contour(X,Y,Z, levels = level_map, cmap = cm.cool)

'''Agregamos el color bar'''
fig2.colorbar(cp)
```
